# Radial Transformations

These one-dimensional grids can be transformed to other kinds of intervals using the
{func}`radial transform <grid.rtransform>` module.

| Transform | Domain | Co-Domain | $r(x)$ |
|-----------|--------|-----------|--------|
| {func}`BeckeRTransform<grid.rtransform.BeckeRTransform>` | $[-1,1]$ | $[r_{min},\infty)$ | $r_i = R \frac{1 + x_i}{1 - x_i} + r_{min}$ |
| {func}`LinearFiniteRTransform<grid.rtransform.LinearFiniteRTransform>` | $[-1,1]$ | $[r_{min},r_{max}]$ | $r_i = \frac{r_{max} - r_{min}}{2} (1 + x_i) + r_{min}$ |
| {func}`InverseRTransform<grid.rtransform.InverseRTransform>` | Co-Domain | Domain | $x_i(r_i)$ |
| {func}`IdentityRTransform<grid.rtransform.IdentityRTransform>` | Domain | Domain | $x_i$ |
| {func}`LinearInfiniteRTransform<grid.rtransform.LinearInfiniteRTransform>` | $[0,\infty)$ | $[r_{min},r_{max})$ | $r_i = \frac{r_{max} - r_{min}}{N - 1} x_i + r_{min}$ |
| {func}`ExpRTransform<grid.rtransform.ExpRTransform>` | $[0,\infty)$ | $[r_{min},r_{max}]$ | $r = r_{min} e^{x \log\left(\frac{r_{max}}{r_{min}} / (N - 1)\right)}$ |
| {func}`PowerRTransform<grid.rtransform.PowerRTransform>` | $[0,\infty)$ | $[r_{min},r_{max}]$ | $r = r_{min} (x + 1)^{\frac{\log(r_{max}) - \log(r_{min})}{N}}$ |
| {func}`HyperbolicRTransform<grid.rtransform.HyperbolicRTransform>` | $[0,\infty)$ | $[0,\infty)$ | $r_i = \frac{a x_i}{1 - bx_i}$ |
| {func}`MultiExpRTransform<grid.rtransform.MultiExpRTransform>` | $[-1,1]$ | $[r_{min},\infty)$ | $r_i = -R \log\left(\frac{x_i + 1}{2}\right) + r_{min}$ |
| {func}`KnowlesRTransform<grid.rtransform.KnowlesRTransform>` | $[-1,1]$ | $[r_{min},\infty)$ | $r_i = r_{min} - R \log\left(1 - 2^{-k}(x_i + 1)^k\right)$ |
| {func}`HandyRTransform<grid.rtransform.HandyRTransform>` | $[-1,1]$ | $[r_{min},\infty)$ | $r_i = R \left(\frac{1+x_i}{1-x_i}\right)^m + r_{min}$ |
| {func}`HandyModRTransform<grid.rtransform.HandyModRTransform>` | $[-1,1]$ | $[r_{min},r_{max}]$ | $r_i = \frac{(1+x_i)^m (r_{max} - r_{min})}{2^m(1 - 2^m + r_{max} - r_{min}) - (1+x_i)^m(r_{max}-r_{min}-2^m)} + r_{min}$ |

# One-Dimensional Grids

There are various choices of one dimensional grids for integrating functions over some finite or infinite
intervals.

## One-Dimensional Quadrature Grids With Explicit Solutions

| Grid | Domain | $x_i$ | $w_i$ |
|------|--------|-------|-------|
| {func}`UniformInteger(Horton2)<grid.onedgrid.UniformInteger>` | $[0,N]$ | $i - 1$ | $1$ |
| {func}`Gauss-Chebyshev Type2<grid.onedgrid.GaussChebyshevType2>` | $[-1,1]$ | $\cos\left( \frac{i}{n+1} \pi \right)$ | $\frac{\pi}{n+1} \sin^2 \left( \frac{i}{n+1} \pi \right)$ |
| {func}`Gauss-Chebyshev Lobatto<grid.onedgrid.GaussChebyshevLobatto>` | $[-1,1]$ | $\cos\left( \frac{(i-1)}{n-1}\pi \right)$ | $w_{1} = w_{n} = \frac{\pi}{2(n-1)}, \quad w_{i\neq 1,n} = \frac{\pi}{n-1}$ |
| {func}`Trapezoidal Lobatto<grid.onedgrid.Trapezoidal>` | $[-1,1]$ | $-1 + 2 \left(\frac{i-1}{n-1}\right)$ | $w_1 = w_n = \frac{1}{n}, \quad w_{i\neq 1,n} = \frac{2}{n}$ |
| {func}`Rectangle-Rule Sine End Points<grid.onedgrid.RectangleRuleSineEndPoints>` | $[-1,1]$ | $\frac{i}{n+1}$ | $\frac{2}{n+1} \sum_{m=1}^n \frac{\sin(m \pi x_i)(1-\cos(m \pi))}{m \pi}$ |
| {func}`Rectangle-Rule Sine<grid.onedgrid.RectangleRuleSine>` | $[-1,1]$ | $\frac{2 i - 1}{2 N_{pts}}$ | $\frac{2}{n^2 \pi} \sin(n\pi x_i) \sin^2(n\pi /2) + \frac{4}{n \pi} \sum_{m=1}^{n-1} \frac{\sin(m \pi x_i)\sin^2(m\pi /2)}{m}$ |
| {func}`Tanh Sinh<grid.onedgrid.TanhSinh>` | $[-1,1]$ | $\tanh\left( \frac{\pi}{2} \sinh(i\delta) \right)$ | $\frac{\frac{\pi}{2}\delta \cosh(i\delta)}{\cosh^2(\frac{\pi}{2}\sinh(i\delta))}$ |
| {func}`Simpson<grid.onedgrid.Simpson>` | $[-1,1]$ | $-1 + 2 \left(\frac{i-1}{N_{pts}-1}\right)$ | $w_i = 2 / (3(N - 1))$ if $i = 0$, $8 / (3(N - 1))$ if $i \geq 1$ and odd, $4 / (3(N - 1))$ if $i \geq 2$ and even |
| {func}`MidPoint<grid.onedgrid.MidPoint>` | $[-1,1]$ | $-1 + \frac{2i + 1}{N_{pts}}$ | $\frac{2}{n}$ |
| {func}`Clenshaw-Curtis<grid.onedgrid.ClenshawCurtis>` | $[-1,1]$ | $\cos (\pi (i - 1) / (N_{pts} - 1))$ | complex weight formula |
| {func}`Fejer First<grid.onedgrid.FejerFirst>` | $(-1,1)$ | $\cos\bigg(\frac{(2i - 1)\pi}{2N_{pts}}\bigg)$ | complex weight formula |
| {func}`Fejer Second<grid.onedgrid.FejerSecond>` | $(-1,1)$ | $\cos(i \pi / N_{pts})$ | complex weight formula |

Grid also includes popular quadrature grids that can integrate polynomials up to degree $2n - 1$:

- {func}`Gauss-Legendre<grid.onedgrid.GaussLegendre>` (domain $[-1, 1]$),
- {func}`Gauss-Chebyshev<grid.onedgrid.GaussChebyshev>` (domain $[-1, 1]$),
- {func}`Gauss-Laguerre<grid.onedgrid.GaussLaguerre>` (domain $[0, \infty)$).

The Trefethen polynomial transformation of various grids on $[-1, 1]$ is also included:

- {func}`Trefethen (Clenshaw-Curtis)<grid.onedgrid.TrefethenCC>`,
- {func}`Trefethen (Gauss-Chebyshev)<grid.onedgrid.TrefethenGC2>`,
- {func}`Trefethen (General)<grid.onedgrid.TrefethenGeneral>`,
- {func}`Trefethen Strip (Clenshaw-Curtis)<grid.onedgrid.TrefethenStripCC>`,
- {func}`Trefethen Strip (Gauss-Chebyshev)<grid.onedgrid.TrefethenStripGC2>`,
- {func}`Trefethen Strip (General)<grid.onedgrid.TrefethenStripGeneral>`.

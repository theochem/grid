# Spherical Integration with Maximum Determinant Points

This example demonstrates how to use the `MaxDeterminantGrid` class for numerical integration on the unit sphere.

## 1. Introduction
Maximum determinant points (Fekete points) provide a stable set of points for integration on the sphere with positive weights. 
For a given degree $t$, there are $N = (t+1)^2$ points.

```python
import numpy as np
from grid.max_det import MaxDeterminantGrid
from grid.utils import generate_real_spherical_harmonics

# Create a Max-Det grid of degree 5
grid = MaxDeterminantGrid(degree=5)
print(f"Grid Size: {grid.size}")
print(f"Degrees Supported: {grid.degree}")
```

## 2. Integration of a simple function
Let's integrate $f(x, y, z) = x^2 + y^2 + z^2$ over the unit sphere. 
Since $x^2 + y^2 + z^2 = 1$ on the sphere, the integral should be $4\pi$.

```python
def func(p):
    return p[:, 0]**2 + p[:, 1]**2 + p[:, 2]**2

integral = grid.integrate(func)
print(f"Calculated Integral: {integral}")
print(f"Expected Integral:   {4 * np.pi}")
print(f"Error:               {np.abs(integral - 4 * np.pi)}")
```

## 3. Integration of Spherical Harmonics
Spherical harmonics of degree $l > 0$ should integrate to 0.

```python
r = np.linalg.norm(grid.points, axis=1)
phi = np.arccos(grid.points[:, 2] / r)
theta = np.arctan2(grid.points[:, 1], grid.points[:, 0])

l = 2; m = 0
sph_harm = generate_real_spherical_harmonics(l, theta, phi)
Y_lm = sph_harm[l**2, :] # Horton 2 index for (l=2, m=0)

integral_Y = grid.integrate(Y_lm)
print(f"Integral of Y_{l}{m}: {integral_Y}")
```
